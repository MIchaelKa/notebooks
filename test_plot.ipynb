```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(2,1))
ax.scatter(range(10), range(10))
```

```python
b = np.array([1,1,0,0,0,0,0,0,1,1] * 200)

plt.figure(figsize=(12,5))
# lw - linewidth
plt.scatter(range(len(b)), [5] * len(b), c=b, marker='_', lw=20)
```

```python
plt.figure(figsize=(12,5))
# lw - linewidth
plt.scatter(range(len(b)), [15] * len(b), c=b, marker='_', lw=20)
```
