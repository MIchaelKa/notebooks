```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# make data
t_max = 24 - 1 # sec
t = np.linspace(0, t_max, 24)
# same as np.arange(24))

angular_velocity = np.pi / 6
R = 1 # m
velocity = angular_velocity * R

angle = angular_velocity * t
angle += -np.pi /2

z = np.cos(angle)
x = np.sin(angle) + 1

plt.figure(figsize=(5, 5))
# plt.plot(z, x)
plt.scatter(z, x)
```

## Скорость равна скорости света

В верхней точке мы практически стоим на месте так как электрон движется в одну сторону со скоростью с, а график двигается в другую. 

При этом график движение является *гипоциклоидой*

```python
c = velocity
ct = t*c+z

plt.figure(figsize=(20,5))

fig, ax = plt.subplots(1, 2, figsize=(20, 5), gridspec_kw={'width_ratios': [1, 3]})

ax[0].scatter(z, x)
ax[1].scatter(ct, x)
```

## Скорость света в 5 раз больше

```python
c = velocity * 5
ct = t*c+z
n = np.arange(len(t))

plt.figure(figsize=(20,5))

fig, ax = plt.subplots(1, 2, figsize=(20, 5), gridspec_kw={'width_ratios': [1, 3]})

ax[0].scatter(z, x, c=n)
ax[1].scatter(ct, x, c=n)
```

## Определяем кажушееся движение

```python
c = velocity * 2
ct = t*c+z
```

```python
# время которое мы знаем
T = ct[5] / c
T
```

```python
angle = angular_velocity * T
angle += -np.pi /2
z_T = np.cos(angle)
x_T = np.sin(angle) + 1
z_T, x_T
```

```python
tau = 5 # время которое мы не знаем
t[tau]
```

```python
# значение в ct[tau] соответствует тому tau которое мы хотим найти
T * c, ct[tau], t[tau] * c + z[tau]
```

```python
plt.figure(figsize=(20,5))

fig, ax = plt.subplots(1, 2, figsize=(20, 5), gridspec_kw={'width_ratios': [1, 3]})

ax[0].scatter(z, x)
ax[1].scatter(ct, x)

ax[0].scatter(z_T, x_T)
```

## Эффект Доплера
Движущийся осциллирующий источник

Пусть осциллирующий заряд двигается по направлению к наблюдателю

```python
# make data
t_max = 24 - 1 # sec
t = np.linspace(0, t_max, 24)
# same as np.arange(24))

angular_velocity = np.pi / 6
R = 1 # m
velocity = angular_velocity * R

angle = angular_velocity * t
angle += -np.pi /2

z = np.cos(angle)
x = np.sin(angle) + 1
```

```python
plt.figure(figsize=(20,5))

fig, ax = plt.subplots(1, 2, figsize=(20, 5), gridspec_kw={'width_ratios': [1, 3]})

ax[0].scatter(z, x)
ax[1].scatter(t, x)
```

```python
n = np.arange(len(t))
```

```python
# движущийся источник
velocity = -1 # отрицательная скорость так как двигаемся ПО направлению к наблюдателю
z_1 = t * velocity

plt.figure(figsize=(10, 5))
plt.scatter(z_1, x, c=n)
```

```python
c = 4
ct = t*c+z_1

plt.figure(figsize=(10, 5))
plt.scatter(ct, x, c=n)
```
